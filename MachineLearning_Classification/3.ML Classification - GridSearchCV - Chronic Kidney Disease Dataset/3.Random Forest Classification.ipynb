```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
               'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
               'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
               'appet_yes', 'pe_yes', 'ane_yes']]
dep=dataset["classification_yes"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
indep = sc.fit_transform(indep)
```

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':['gini', 'entropy'],
              'max_features': ['sqrt','log2'],
              'n_estimators':[10,100]}

grid = GridSearchCV(RandomForestClassifier(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1')

grid.fit(indep, dep)
```

```python
re=grid.cv_results_

grid_predictions = grid.predict(indep)

table=pd.DataFrame.from_dict(re)
table
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(dep, grid_predictions)    
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(dep, grid_predictions)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(dep,grid.predict_proba(indep)[:,1])
```
