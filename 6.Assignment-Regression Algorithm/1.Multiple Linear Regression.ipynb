```python
#importing the lib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the dataset
dataset=pd.read_csv('insurance_pre.csv') 
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
indep = dataset[['age', 'sex_male', 'bmi', 'children', 'smoker_yes']]

dep = dataset[['charges']]
```

```python
indep
```

```python
dep
```

```python
#split into training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(indep, dep, test_size=1/3, random_state=0)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train) # y=w*x1+b0 for this equation we will get the value for b1 and b0
```

```python
#Viewing b1 and b0 value
weight=regressor.coef_
print("Weight of the model={}".format(weight))

bais=regressor.intercept_
print("Intercept of the model={}".format(bais))
```

```python
y_pred=regressor.predict(x_test)
```

```python
#https://scikit-learn.org/stable/modules/model_evaluation.html
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
